```python
## while loop

## The while loop continues to execute as long as a certain condition is true

count = 0

while count < 5:
    print(count)
    count = count + 1
```

```python
count = 0


while count % 2 == 0:
    print(count)
    count = count + 1
```

```python
## Loop control statements

## break
## The break statement exit the loop permaturely


## break statement

for i in range (10):
    if i == 5:
        break
    print(i)
```

```python
## Continue

## The continue statement skips the current iteration and continue to the next.

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
## pass

## The pass statement is a null operation; it does nothing

for i in range(5):
    if i  == 3:
        print("The number is", i)
        pass
    print(i)
```

```python
## Nested loops
## A loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples - Calculate the sum of first N natural numbers using while and for loop

## While loop

n = 10
sum = 0
count = 1

while count <= n:
    sum = sum + count
    count = count + 1

print("Sum of firs", n , "natural number is: ", sum)
```

```python
## for loop

n = 10
sum = 0
count = 1
for i in range(11):
    sum = sum + i

print(sum)
```

```python
## Example - Prime numbers between 1 and 100

for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```

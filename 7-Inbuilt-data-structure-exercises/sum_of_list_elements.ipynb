```python
'''
BEGIN
GET the list of elements
SET a certain total = 0
FOR each element in the list, we add it to total
PRINT total
END
'''
```

```python
def sum_list_of_element(numbers):
    """
    Function returns the sum of elements in a list

    Parameter:
    numbers (int) : The list of numbers

    Returns:
    float/float : sum of elements in the list
    """

    # initialize a certain total
    total = 0

    # loop through the list from start to end and add each element to total

    for num in numbers:
        total = total + num
    
    return total

print(sum_list_of_element([10, 20, 30, 40]))
```

```python
# Alternatively, Making use of the sum() method on the list

def sum_of_list_elements(numbers):
    return sum(numbers)

print(sum_of_list_elements([10, 20, 30, 40]))
```

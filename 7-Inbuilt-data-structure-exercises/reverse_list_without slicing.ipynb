```python
"""
START
INPUT the list
FOR each item the list
SET end to point to the end of the list and start to point to the start of the list and then swap
RETURN the reversed list
END

"""
```

```python
def reverse_list(lst):
    # Initialize two point, one pointing to the begining and the other to the end
    start = 0
    end = len(lst) - 1

    # Swap elements until the two pointers reach at the middle
    while start < end:
        # Swap the elements at the the end and start pionters
        lst[start], lst[end] = lst[end], lst[start]
        # Move the pointers towards the middle
        start += 1
        end -= 1
    # Return our reversed list
    return lst

print(reverse_list([1, 2, 3, 4, 5]))
```

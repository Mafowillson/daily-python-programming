```python
"""
START
GET the list of numbers
SET the the lagert element to be the first element in the list
FOR all the numbers in the list, compare it to the largest element
IF any number is greater than our current largest element, we set the element to be lagert element
RETURN largest element
END
"""
```

```python
def largest_element_in_a_list(numbers):
    """
    Function returns the largest element in a list

    Parameter:
    numbers (list) : An iterable of numbers

    Returns:

    number (int / float): Which is the largest element in the list
    """
    if not numbers:
        return None

    # seting largest element to be first element in the array
    largest = numbers[0]

    for number in numbers:
        if number > largest:
            largest = number
    
    return largest

print(largest_element_in_a_list([30, 10, 20, 40, 1]))
```

```python
def find_largest(numbers):
    # Check if the list is empty
    if not numbers:
        return None # or raise an exception, depending on requirements
    
    # Initialize the largest number as the first element
    largest = numbers[0]

    # Iterate through the list starting from the second element
    for num in numbers[1:]:
        # If we find a number larger than our current largest, update largest
        if num > largest:
            largest = num
    # Return the largest number found
    return largest

print(find_largest([2, 3, 50, 100, 4, 5, 700]))
```

```python
# Alternatively, using the max function and list comprehension

def find_largest(numbers):
    return max(numbers) if numbers else None

print(find_largest([50, 60, 5, 1, 29]))
```

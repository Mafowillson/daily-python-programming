```python
"""
START
GET the list
INITIALIZE an empty list to store unique element
FOR each element in the list we store it in our unique list if not already in the list
IF num is in unique list, we retrun False else True
END
"""
```

```python
def are_element_unique(lst):
    unique_el = []
    for num in lst:
        if num in unique_el:
            return False
        unique_el.append(num)
    return True

print(are_element_unique([1, 2, 2, 3]))
print(are_element_unique([1, 2, 3]))
        
```

```python
def check_unique(lst):
    # Initial an empty set to keep track of seen elements
    seen = set()

    # Iterate through each element in the input list
    for num in lst:
        # If the element has been seen before, return false
        if num in seen:
            return False
        # Add the element to the seen set
        seen.add(num)

    # If no duplicates are found, return True
    return True

print(check_unique([1, 2, 3, 4, 4]))
print(check_unique([1, 2, 3, 4, 5, 6]))
```

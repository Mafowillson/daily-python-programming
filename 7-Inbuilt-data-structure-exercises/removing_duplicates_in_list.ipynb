```python
"""
START
GET the list
INITIALIZE an empty list to store unique element
FOR each element in the list we store it in our unique list if not already in the list
RETURN unique list
END
"""
```

```python
lst = [1, 2, 2, 3, 4, 4, 5]
unique_list = []

for num in lst:
    if num in unique_list:
        continue
    else:
        unique_list.append(num)

print(unique_list)
```

```python
def remove_duplicates(lst):
    unique_list = []
    for item in lst:
        # Check if item is already in the unique_list
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

print(remove_duplicates([1, 2, 2, 3, 4, 4, 5]))
```

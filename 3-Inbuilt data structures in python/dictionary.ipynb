```python
## Creating dictionaries

empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name": "Mafo willson", "age": 32, "grade": 24}
print(student)
print(type(student))
```

```python
# Single key is always used

student = {"name": "Mafo willson", "age": 32, "name": 24}
print(student)
```

```python
## accessing dictionary elements

student = {"name": "Mafo willson", "age": 32, "grade": 'A'}
print(student)
```

```python
print(student['grade'])
print(student['age'])
## Accessing using the get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', "Not available"))
```

```python
## Modifying Dictionary Elements

## Dictionary are mutable, so you can add, update or delete elements

print(student)
```

```python
student['age'] = 33 ## update value for a key
print(student)
student["address"] = "India" ## Added a new key and value
print(student)
```

```python
del student['grade'] ## deletes the key and value pair

print(student)
```

```python
## Dictionay methods

keys = student.keys() ## get all keys

print(keys)

values = student.values() ## get all values

print(values)

items = student.items() ## get all key value pairs

print(items)
```

```python
## Shallow copy

student_copy = student
print(student)
print(student_copy)
```

```python
student['name'] = "Willson2"
print(student)
print(student_copy)
```

```python
## shallow copy

student_copy1 = student.copy() ## shallow copy

print(student_copy1)
print(student)
```

```python
student["name"] = "Willson3"
print(student_copy1)
print(student)
```

```python
## Iterating over dictionaries
## You can use loops to iterate over dictionaries, keys, values, or items

## Iterating over keys

for keys in student.keys():
    print(keys)

```

```python
## Iterating over values

for value in student.values():
    print(value)
```

```python
## Iterating over key value pairs

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested Dictionaries

students = {
    "student1": {
        "name": "Willson",
        "age": 32
    },
    "student2":{
        "name": "Peter",
        "age":35
    }
}
print(students)
```

```python
## Access nested dictionary elemets
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
# Iterating over nested dictionaries

for student_id, student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key, value in student_info.items():
        print(f"{key} : {value}")
```

```python
## Dictionary comprehension

squares = {x : x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension

evens = {x : x**2 for x in range(10) if x % 2 == 0}
print(evens)
```

```python
## Practical examples

## Use a dictionary to count the frequency of elements in list

numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
## key word aguement (**x) are values that occurs as key value pairs
```

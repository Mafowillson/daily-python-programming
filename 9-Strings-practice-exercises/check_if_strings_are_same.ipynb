```python
def are_equal_strings(s, t):
    """
    Function to chech if two strings are equal without using built-in functions.

    Parmeters:
    s (str): First string
    t (str): Second string

    Returns:
    bool: True if the strings are equal, False otherwise.
    """

    len_s = 0
    len_t = 0

    # Calculate the length of the string manually
    while len_s < len(s):
        len_s += 1
    while len_t < len(t):
        len_t += 1
    
    # If lengths are not equal, the strings can't be equal
    if len_s != len_t:
        return False
    
    # Compare characters one by one
    for i in range(len_s):
        if s[i] != t[i]:
            return False # If a character does not match, return false
        
    return True # If all character match, return true

print(are_equal_strings('Hello', 'hello'))
```

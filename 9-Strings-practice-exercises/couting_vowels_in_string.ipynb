```python
"""
START
GET the string
INITIALIZE vowel count to 0
STORE all possible in string know as vowels both upper and lower case
FOR every character in the string if there exist a vowel, we increment vowel count
RETURN vowel count
STOP
"""
```

```python
s = input("Enter the string: ")
vowels = 'aeiouAEIOU'
vowel_count = 0

for i in range(len(s)):
    for vowel in vowels:
        if s[i] == vowel:
            vowel_count += 1

print(vowel_count)
```

```python
def count_wowels(s):
    
    """
    Function to count the number of vowels in the input string

    Parameters:
    s (str): The input string to check for vowels

    Returns: 
    int: The count of vowels in the input string. 
    """
    # Define the set of vowels (both lowercase and uppercase)
    vowels = "aeiouAEIOU"

    # Initialize a counter for vowels
    count = 0

    # Loop through each character in the string
    for char in s:
        # Check if the character is a vowel
        if char in vowels:
            count += 1 # Increment the count if it is a vowel
    # Return the total count of vowels
    return count

print(count_wowels('Hello, World!!'))
```

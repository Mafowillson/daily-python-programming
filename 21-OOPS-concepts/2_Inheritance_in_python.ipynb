Inheritance in Python

Inheritance is a fundamental concept in OOP that allows a class to inherit attributes and methods from another class. THis lesson covers single inheritance and multiple inheritance, demonstrating  how to create and use them in python


```python
# Inheritance (Single inheritance)
## Parent class

class Car:
    def __init__(self, windows, doors, engintype):
        self.windows = windows
        self.doors = doors
        self.engintype = engintype
    
    def drive(self):
        print(f"The person will drive the {self.engintype} car")
```

```python
car1 = Car(4, 5, "petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, engintype, is_selfdriving):
        super().__init__(windows, doors, engintype)
        self.is_selfdriving = is_selfdriving
    
    def selfdriving(self):
        print(f"Tesla suppors self driving: {self.is_selfdriving}")
```

```python
tesla1 = Tesla(4, 4, "diesel", True)
tesla1.drive()
tesla1.selfdriving()
```

```python
## Multiple Inheritance
## When a class inherits from one base class

## Base class1 
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("Subclass must implement this method")

## Base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

## Derived class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
    
    def speak(self):
        return f"{self.name} says woof"

## Create an object
dog = Dog("Buddy", "Willson")
print(dog.speak())
print(f"Owner: {dog.owner}")
    
```

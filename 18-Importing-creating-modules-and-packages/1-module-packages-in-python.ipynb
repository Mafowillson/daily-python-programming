Importing modules in python: Modules and packages

In python, modules and packages help organize and reuse code. Here's a comphrehensive guide on how to import them.

```python
import math

math.sqrt(16)
```

```python
from math import sqrt, pi
print(sqrt(16))
print(pi)
```

```python
import numpy as np
np.array([1, 2, 3, 4])
```

```python
from math import *
print(sqrt(16))
print(pi)
print(sin(-1))
```

```python
from package.maths import addition

print(addition(2, 3))
```

```python
from package import maths

maths.addition(2, 3)
```

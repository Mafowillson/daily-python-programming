```python
'''
Psuedo coded for pyramid pattern

GET the number of rows for the pyramid
FOR i -> 1, n
PRINT n - i spaces, and 2 * i - 1 stars

'''
```

```python
rows = int(input("Enter the number of stars: "))

for i in range(1, rows + 1):
    print(" " * (rows - i), "*" * (2 * i - 1))
```

```python
def generate_pyramid(n):
    """
    Function to return a pyramid of '*' of side n as a list of strings

    parameters:
    n (int): The number of rows in the pyramid

    Returns:
    list: A list of strings where each string represents a row in the pyramid
    """
    # Initialize an empty list to store the rows of the pyramid

    pyramid = []

    # Loop through each row from 1 to n
    for i in range(1, n + 1):
        # Number of stars in the current row is 2 * i - 1
        stars = '*' * (2 * i -1)
        # Number of leading spaces to center the stars is i - 1
        spaces = '' * (n - i)
        # Add the centerd row to the list
        pyramid.append(spaces + stars + spaces)
    
    # Return the list of pyramid rows
    return pyramid

print(generate_pyramid(5))
```

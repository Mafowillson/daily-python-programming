```python
side = int(input("Enter the side of the square: "))


for i in range(side):
    for j in range(side):
        print('*', end=" ")
    print()
```

```python
def generate_squre(n):
    """
    Function to return a square pattern of '*' of side n as a list of strings. 

    parameters:

    n (int): Then size of the square.

    returns:

    list: A list of strings where each string represents a row of the square.

    """
    # Initialize an empty list to store the rows of the square

    square = []

    # loop n times to generate each row 
    for i in range(n):
        # create a row of "*" of length n and append it to the list
        square.append('*' * n)
    
    # return list of rows
    return square

# print(generate_squre(5))

size = int(input("Enter size of the square"))

for row in generate_squre(size):
    print(row)
```

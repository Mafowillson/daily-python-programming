```python
"""
Pseudo code for reversed pyramid

BEGIN
GET the number of row for which you want to generate the pyramid
FOR i -> (number of rows, 1)
PRINT -> n - i to generate the spaces
PRINT -> 2 * i - 1 to generate the *
End
"""
```

```python
rows = int(input("Enter the number of rows: "))

for i in range(rows, 0, -1):
    print(" " * (rows - i), "*" * (2 * i - 1))
```

```python
def generate_inverted_pyramid(n):
    """
    Function to retrun an inverted pyramid of '*' of side n as a list of strings

    parameter:
    n (int): The number of rows for the inverted pyramid

    Return: 
    list: A list of strings where each string represents a row of the inverted pyramid
    """

    # Initialize an empty list to store the rows of the pyramid
    pyramid = []
    
    # Loop through each row from 1 to n
    for i in range(1, n + 1):
        # Number of stars i the current row is 2 * (n - i + 1) - 1
        stars = '*' * (2 * (n - i + 1) - 1)
        # Numner of leading spaces to center the star is i - 1
        spaces = '' * (n - i)
        # Add the centered row to the list
        pyramid.append(spaces + stars + spaces)

    # Return the list of pyramid rows
    return pyramid

print(generate_inverted_pyramid(5))
```

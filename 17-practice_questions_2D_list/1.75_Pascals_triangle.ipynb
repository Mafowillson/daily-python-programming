```python
def pascal_triangle(rows):
    # Initialize and empty triangle to store the rows of the pascals triangle
    triangle = []
    for row in range(1, rows):
        for column in range(row):
            # Enure the first and last element of each row of our pascal triangle is 1
            if column == 0 or column == row:
                row[column] = 1
            # For the middle part of each row
            else:
                row[column] = triangle[row-1][column - 1] + triangle[row-1][column]
        # Append the row to the triangle
        triangle.append(row)
    # Return the triangle containing the pascal triangle
    return triangle

print(pascal_triangle(5))
```

```python
def generate(numRows):
    """ 
    Function to generate the first numRows of Pascal's triangle.
    :param numRows: int -> Number of rows of Pascal's triangle to generate
    :return: List[List[int]] -> The first numRows of Pascal's triangle

    # Initialize the pascal's triangle with empty list
    """
    # Initialize the pascal's triangle with an empty list
    triangle = []

    # Generate each row of the triangle
    for row_num in range(numRows):
        # Start each row a list of 1's 
        row = [1] * (row_num + 1)

        # Update the values of the row based on the previous row
        for j in range(1, row_num):
            row[j] = triangle[row_num - 1][j - 1] + triangle[row_num - 1][j]

        # Append the generated row to the triangle
        triangle.append(row)

    return triangle

for row in generate(5):
    for column in row:
        print(column, end=" ")
    print("")
```
